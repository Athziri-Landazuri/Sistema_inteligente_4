```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

Archivo: pop.csv

```python
df = pd.read_csv('pop.csv') # 1,2,3
print(df)
```

```python
x = np.asanyarray(df[['year']])
y = np.asanyarray(df[['pop']])
```

```python
plt.figure()
plt.grid()
plt.title('Regresion lineal')
plt.xlabel('year')
plt.ylabel('poblacion')

plt.plot(x,y,'bo')

plt.show()
```

```python
model = linear_model.LinearRegression()
model.fit(x,y)

print('Coeficientes(',model.intercept_,',',model.coef_,')')
```

```python
yp = model.predict(x)
xp_1 = np.array([[2003]]);
yp_1 = model.predict(xp_1)
```

```python
plt.figure()
plt.grid()
plt.title('Regresion lineal')
plt.xlabel('year')
plt.ylabel('poblacion')

plt.plot(x,y,'bo')
plt.plot(x,yp,'r-',lw=2)
plt.plot(xp_1,yp_1,'gs',lw=2)

plt.legend(['muestras','prediccion','(xp,yp)'])
plt.show()
```

```python
e1 = mean_absolute_error(y,yp)
print('MAE =',e1)

e2 = mean_squared_error(y,yp)
print('MSE =',e2)

r2 = r2_score(y, yp)
print('R2 =',r2)
```
